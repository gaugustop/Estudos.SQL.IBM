## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table


```python
# Rows in Crime table
cursor.execute('select count(distinct("ID")) from "CRIME"')
print(cursor.fetchall()[0][0])
```

### Problem 2

##### Retrieve first 10 rows from the CRIME table


```python
pd.read_sql('select * from "CRIME" limit 10', con)
```

### Problem 3

##### How many crimes involve an arrest?


```python
cursor.execute('select count(*) from "CRIME" where "ARREST" = 1')
print(cursor.fetchall()[0][0])
```

### Problem 4

##### Which unique types of crimes have been recorded at GAS STATION locations?


```python
cursor.execute('select distinct("PRIMARY_TYPE") from "CRIME" '
               'where "LOCATION_DESCRIPTION" = "GAS STATION"')
for line in cursor.fetchall():
    print(line[0])
```

Hint: Which column lists types of crimes e.g. THEFT?


### Problem 5

##### In the CENUS_DATA table list all Community Areas whose names start with the letter ‘B’.


```python
cursor.execute('select "COMMUNITY_AREA_NAME" from "CENSUS" '
                'where "COMMUNITY_AREA_NAME" like "B%"')
for line in cursor.fetchall():
    print(line[0])
```

### Problem 6

##### Which schools in Community Areas 10 to 15 are healthy school certified?


```python
cursor.execute('select "NAME_OF_SCHOOL" from "SCHOOLS" ' 
                'where "COMMUNITY_AREA_NUMBER" <= 15 and "COMMUNITY_AREA_NUMBER" >=10 ' 
                'and "HEALTHY_SCHOOL_CERTIFIED" = "Yes"')
for line in cursor.fetchall():
    print(line[0])
```

### Problem 7

##### What is the average school Safety Score?


```python
cursor.execute('select avg("SAFETY_SCORE") as "AVERAGE_SAFETY_SCORE" from "SCHOOLS" ')
cursor.fetchall()[0][0]
```

### Problem 8

##### List the top 5 Community Areas by average College Enrollment \[number of students]


```python
cursor.execute('select "COMMUNITY_AREA_NAME", avg("COLLEGE_ENROLLMENT") as "AVG_COLLEGE_ENROLLMENT" from "SCHOOLS" '
                'group by "COMMUNITY_AREA_NAME" '
                'order by "AVG_COLLEGE_ENROLLMENT" desc limit 5')
for line in cursor.fetchall():
    print(line[0])
```

### Problem 9

##### Use a sub-query to determine which Community Area has the least value for school Safety Score?


```python
cursor.execute('select "COMMUNITY_AREA_NAME"from "SCHOOLS" '
                'where "SAFETY_SCORE" = (select min("SAFETY_SCORE") from "SCHOOLS") ')
cursor.fetchall()[0][0]
```

### Problem 10

##### \[Without using an explicit JOIN operator] Find the Per Capita Income of the Community Area which has a school Safety Score of 1.


```python
cursor.execute('select C."PER_CAPITA_INCOME "  from "CENSUS" C, "SCHOOLS" S '
               'where upper(C."COMMUNITY_AREA_NAME") = upper(S."COMMUNITY_AREA_NAME") and  S."SAFETY_SCORE" = 1')
cursor.fetchall()[0][0]
```
